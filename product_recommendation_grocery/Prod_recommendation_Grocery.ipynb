```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_2.head(10)
```

```python
for i in df_2.iloc[:,1:].columns:
    print(df_2[i].unique())
```

```python
df_2=df_2.astype('int64')
```

```python
df_2.set_index('SI_no',inplace=True)
```

```python
#applying apriori algorithm to generate rules which have support more than 0.05
frequent_itemsets = apriori(df_2, min_support=0.05, use_colnames=True)
frequent_itemsets.head()
```

```python
#calculating confidence and lift for the genrated rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules.sort_values(['confidence'],ascending=False)
```

```python
rules.sort_values(['support'],ascending=False)
```
