```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
%matplotlib inline
```

```python
train_df1 = df1.dropna(subset=['Title'])
```

```python
train_df1=train_df1.dropna(subset=['Review Text'])
```

```python
train_df1.loc[train_df1["Division Name"].isnull()]
```

```python
train_df1=train_df1.dropna(subset=['Division Name'])
```

```python
train_df1.shape
```

```python
train_df1.describe()
```

```python
#Creating new Rating Class 
train_df1['Rating Class']=pd.Series(index=train_df1.index)
```

```python
#Rating Class = bad if Rating is < or = 3
train_df1.loc[train_df1['Rating']<=3,'Rating Class']='Bad'
```

```python
#Rating Class = good if Rating is > 3
train_df1.loc[train_df1['Rating']>3,'Rating Class']='Good'
```

```python
train_df1.reset_index(inplace=True)
```

```python
train_df1.head()
```

```python
train_df1=train_df1.iloc[:,1:]
```

```python
print(train_df1['Division Name'].unique())
print(len(train_df1['Division Name'].unique()))
print(train_df1['Department Name'].unique())
print(len(train_df1['Department Name'].unique()))
print(train_df1['Class Name'].unique())
print(len(train_df1['Class Name'].unique()))
```

```python
# Categorical encoding for Rating class 
train_df1['Rating Class Code']=train_df1['Rating Class'].replace({'Good':1,'Bad':0})
```

```python
# Categorical encoding for other ojbect variables

from sklearn.preprocessing import LabelEncoder
ld=LabelEncoder()
train_df1['Division Name Code']=ld.fit_transform(train_df1['Division Name'])
train_df1['Department Name Code']=ld.fit_transform(train_df1['Department Name'])
train_df1['Class Name Code']=ld.fit_transform(train_df1['Class Name'])
```

```python
#using decision tree
from sklearn import tree
from sklearn import metrics
```

```python
my_tree_one = tree.DecisionTreeClassifier(criterion="gini", max_depth=3)
my_tree_one = my_tree_one.fit(x_train, y_train)
my_tree_one
```

```python
print(my_tree_one.feature_importances_)
print(my_tree_one.score(x_train, y_train))
list(zip(colx,my_tree_one.feature_importances_))
```

```python
pred = my_tree_one.predict(x_test)

auc=metrics.roc_auc_score(y_test,pred)
auc
```

```python
print("classification_report")
print(metrics.classification_report(y_test,pred))
print("confusion_matrix")
print(metrics.confusion_matrix(y_test,pred))
```
